```python
#convocando as jogadoras com todas as informa√ß√µes necess√°rias
jogadoras = []

for convocadas in range (14):
    nome = input("Nome da jogadora: ").title().strip()
    numero = int(input('N¬∫ da jogadora: '))
    idade = int(input('Idade : '))
    altura = float(input ('Altura: '))
    posicao = input ('Posi√ß√£o: ').title().strip()
    participacao = input ('Participa√ß√£o em mundiais: ').title().strip()
    convocacao = [numero, nome, idade, altura, posicao, participacao]
    jogadoras.append(convocacao)
```

```python
#criando dataframe com as convocadas 
import pandas as pd

df = pd.DataFrame(jogadoras, columns=['N¬∫', 'NOME', 'IDADE', 'ALTURA', 'POSI√á√ÉO', 'PARTICIPA√á√ÉO EM MUNDIAIS'])
df=df.sort_values(by =['N¬∫'])
display(df)

#criando dataframe com m√©dia de altura e idade
media_idade = 0
media_altura = 0
media = []

for nome, numero, idade, altura, posicao, participacao in jogadoras:
    media_idade += idade/14
    media_altura += altura/14
media = [media_idade, media_altura]


dfnew= pd.DataFrame([media], columns=['M√âDIA IDADE', 'M√âDIA ALTURA'] )
display(dfnew.round(2))                                    
```

```python
#contagem de idade e altura
import matplotlib.pyplot as plt

df=pd.DataFrame(jogadoras,columns=['N¬∫', 'NOME', 'IDADE', 'ALTURA', 'POSI√á√ÉO', 'PARTICIPA√á√ÉO EM MUNDIAIS'])
plt.figure(figsize=(15,5))
df['IDADE'].hist(bins=7, rwidth=0.50, color="darkblue")
plt.xlim(17,43) 
plt.ylim(0,5)
plt.title("Mundial 2022 - Idade", fontsize = 16)

plt.figure(figsize=(15,5))
df['ALTURA'].hist(bins=7, rwidth=0.90, color="purple")
plt.xlim(1.65,1.95) 
plt.ylim(0,7)
plt.title("Mundial 2022 - Altura", fontsize = 16)
```

```python
#particapa√ß√µes em mundiais
part_mundiais = df['PARTICIPA√á√ÉO EM MUNDIAIS'].value_counts()
print(part_mundiais)
```

```python
#criando coluna para mostrar novatas e experientes(novatas <= 25)
import sys

mundial['FAIXA ET√ÅRIA']=pd.cut(
    mundial['IDADE'],
    bins=[0, 25, sys.maxsize],
    labels=['Novata', 'Experiente']

)

mundial
```
